# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
print("\n Enter marks:")
marks = int(input())

if marks >= 90:
    print("A")
elif marks >= 80:
    print("B")
elif marks >= 70:
    print("C")
elif marks >= 60:
    print("D")
else:
    print("Fail")
```

# User se day number (1–7) lo:

# 1 → Monday
# 2 → Tuesday
## ...
# 7 → Sunday

```python
print("\nEnter a day number from (1-7) :")
day = int(input())

if day == 1:
    print("Monday")
elif day == 2:
    print("Tuesday")
elif day == 3:
    print("Wednesday")
elif day == 4:
    print("Thursday")
elif day == 5:
    print("Friday")
elif day == 6:
    print("Saturday")
elif day == 7:
    print("Sunday")
else:
    print("Invalid day number")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
print("\nEnter a month number:")
month = int(input())

if month in [12, 1, 2]:
    print("Winter")
elif month in [3, 4, 5]:
    print("Spring")
elif month in [6, 7, 8]:
    print("Summer")
else:
    print("Autumn")
```

# User se 2 numbers lo:

# pehla > dusra → "First is greater"
# pehla < dusra → "Second is greater"

```python
print("\nEnter 2 numbers:")
num1 = int(input("Pehla number: "))
num2 = int(input("Dusra number: "))

if num1 > num2:
    print("First is greater")
elif num1 < num2:
    print("Second is greater")
else:
    print("Both equal")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
print("\nSimple calculator:")
num1 = float(input("First number: "))
num2 = float(input("Second number: "))
operator = input("Operator (+, -, *, /): ")

if operator == "+":
    print(num1 + num2)
elif operator == "-":
    print(num1 - num2)
elif operator == "*":
    print(num1 * num2)
elif operator == "/":
    if num2 != 0:
        print(num1 / num2)
    else:
        print("Cannot divide by zero")
else:
    print("Invalid operator")
```
